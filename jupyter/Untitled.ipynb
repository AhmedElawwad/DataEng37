```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal")

#
plt.title("Fmri time")
plt.xlabel("Time Point")
plt.ylabel("Percent Signal Change")
```

```python
sns.lineplot(
    data= fmri,
    x = "timepoint",
    y= "signal",
    hue = "event"
)
```

```python
sns.lineplot(
    data= fmri,
    x = "timepoint",
    y= "signal",
    hue = "event",
    palette=["teal", "indigo"]
)
```

```python
sns.relplot(
    data= fmri,
    x='timepoint',
    y = 'signal',
    hue = 'event',
    col = "region",
    kind = 'line'
)
```

```python
plt.plot(fmri['timepoint'], fmri['signal'])
```

```python
data = fmri[['timepoint', 'signal']].groupby('timepoint').mean()
```

```python
data
```

```python
plt.plot(data)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.bar(tips.day, tips.total_bill)
```

```python
mean_day = tips[['total_bill', 'day']].groupby('day').mean()
```

```python
plt.bar(mean_day.index, mean_day['total_bill'], color='purple')
```

```python
plt.hist(tips['tip'])
```

```python
sns.barplot(data=tips, x="day", y="total_bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Total Bill amount per day")
plt.show()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True)
plt.figure(figsize=(5,5))
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex')
plt.show()
```
